```python
import pandas as pd
import xml.etree.ElementTree as ET
```

```python
file_path = '../data/Baza_danych_RCN_Poznań_2026.gml'
```

```python
tree = ET.parse(file_path)
root = tree.getroot()
```

```python
# Define the namespace dictionary (required for GML parsing)
ns = {
    'gml': 'http://www.opengis.net/gml/3.2',
    'rcn': 'urn:gugik:specyfikacje:gmlas:rejestrcennieruchomosci:1.0'
}
```

```python
#Extract all notarial deeds (RCN_Dokument) from the XML
document_nodes = root.findall('.//rcn:RCN_Dokument', ns)
print(f"Found notarial deeds: {len(document_nodes)}")
```

```python
document_list = []

# A simple loop to extract tags and their text values
for doc in document_nodes:
    doc_dict = {}
    for child in doc:
        # Strip the long namespace URI from the tag name (everything after '}')
        clean_tag = child.tag.split('}')[-1]
        
        # Extract the text from inside the element (if it exists)
        value = child.text.strip() if child.text else None
        
        doc_dict[clean_tag] = value
        
    document_list.append(doc_dict)

print("\nPreview of the first extracted document:")
print(document_list[0])
```

```python
df_documents = pd.DataFrame(document_list)

# Drop 'ghost' columns that XML might have created as entirely empty (e.g., boundedBy)
df_documents = df_documents.dropna(axis=1, how='all')

print("\nDataFrame Head:")
print(df_documents.head(3))

print("\nData Types:")
print(df_documents.dtypes)
```
